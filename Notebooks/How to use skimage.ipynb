```python
from skimage import io
```

```python
# Ler imagem
img = io.imread('../Extra/sp.jpg')
print(img.shape) # returns a tuple of number of rows, columns and channels if RGB
io.imshow(img);
```

```python
# Salvar imagem
io.imsave('../Extra/new_image.jpg',img)
```

```python
# Converter escalas de cor
from skimage import color

gray_image = color.rgb2gray(img)
print(gray_image.shape) # returns number of rows and columns
io.imshow(gray_image);
```

```python
# Convertendo de rgb para hsv
hsv_image = color.rgb2hsv(img)
print(hsv_image.shape)
io.imshow(hsv_image);
```

```python
# Outra forma de conversao de espaco de cores
new_space_image= color.convert_colorspace(img,'RGB','RGB CIE')
print(new_space_image.shape)
io.imshow(new_space_image);
```

```python
import numpy as np 
from skimage import draw 
```

```python
# Leitura
img = np.zeros((100, 100), dtype=np.uint8)

# Desenha um círculo
x, y = draw.circle(50, 50, 10)
img[x, y] = 1

io.imshow(img);
```

```python
# Cria uma imagem vazia
img = np.zeros((100, 100), dtype=np.uint8)
r = np.array([10, 25, 80, 50]) # Row coordinates of vertices of polygon
c = np.array([10, 60, 40, 10]) # Column coordinates of vertices of polygon

# Desenha um polígono
x, y = draw.polygon(r, c)
img[x, y] = 1


io.imshow(img);
```

```python
# Cria uma imagem vazia
img = np.zeros((100, 100), dtype=np.uint8)

# Desenha um polígono
rr, cc = draw.circle_perimeter(40, 65, 15)

img[rr, cc] = 1


io.imshow(img);
```
